```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Storage.csv')
```

```python
df["name"] = df["Brand"] + ", " + df["Product"]
```

```python
df.head(5)
```

```python
plt.figure(figsize=(20, 10))
for name in df['name'].unique():
    subset = df[df['name'] == name]
    subset = subset.sort_values(by='Size')
    plt.plot(subset['Size'], subset['Price'], label=name, marker='o')

plt.ylabel('Price')
plt.xlabel('Size')
plt.legend(title='name')
plt.xticks(df['Size'])
plt.show()
```
